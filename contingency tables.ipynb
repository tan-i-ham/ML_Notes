# Two-Way Tables
Two-way frequency tables, also called contingency tables, are tables of counts with two dimensions where each dimension is a different variable. Two-way tables can give you insight into the relationship between two variables. To create a two way table, pass two variables to the pd.crosstab() function instead of one:


## One-Way Tables

```python

marital_sex = pd.crosstab(index= df["Marital Status"], 
                           columns= df["Approximated Social Grade"])

marital_sex/marital_sex.sum()


```

## Two-Way Tables


```python
pd.crosstab(index=df["Region"], columns=df["Age"])
```

```python
pd.crosstab(index=df["Region"], columns=df["Age"] , margins=True)
```

```python

a = pd.crosstab(index= df["Health"],columns=df["Religion"], margins = True)
a.columns = ['No code required','No religion','Christian','Buddhist','Hindu','Jewish','Muslim','Sikh','Other religion','Not stated','total']
a


```
